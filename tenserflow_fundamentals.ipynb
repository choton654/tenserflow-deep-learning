```python
random_1 = tf.random.uniform(minval=0, maxval=1, shape=(5,300))
random_2 = tf.random.uniform(minval=0, maxval=1, shape=(5,300)) # create random tensor value between 0 and 1 with shape (5,300)
random_1, random_2
```

```python
multi = tf.math.multiply(random_1, random_2)
multi
```

```python
dot_product = tf.matmul(random_1, tf.reshape(random_2,shape=(300,5)))
dot_product
```
