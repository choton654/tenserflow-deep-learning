```python
nut_butter_prices = {"Almond butter":10,"Peanut butter":8,"Cashew butter":12}
fig, ax = plt.subplots();

ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title='Dans nut butter store', ylabel="Price ($)")
```

```python
car_sales = pd.read_csv('../pandas/car-sales.csv')
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]','')
```

```python
car_sales['Price'] = car_sales['Price'].str[:-2]
```

```python
car_sales
```

```python
car_sales['Sales Date'] = pd.date_range('1/1/2022',periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x='Sales Date', y='Total Sales')
```

```python
car_sales.plot(x='Odometer (KM)', y='Price', kind='scatter');
```
