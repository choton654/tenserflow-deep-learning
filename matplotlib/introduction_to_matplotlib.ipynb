```python
heart_deases = pd.read_csv('heart-disease.csv')
heart_deases.head()
```

```python
heart_deases['age'].plot.hist(bins=10);
```

```python
heart_deases['age'].plot.hist(bins=len(heart_deases));
```

```python
heart_deases.plot.hist(figsize=(10, 30), subplots=True);
```

```python
over_50 = heart_deases[heart_deases['age']  > 50]
over_50.head()
```

```python
over_50.plot(kind='scatter', x='age', y='chol', c='target');
```

```python
fig, ax = plt.subplots()
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
# ax.set_xlim([40,100])
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target']);
ax.set(title='Heart disease and Cholestrol levels', xlabel='age', ylabel='chol')
ax.legend(*scatter.legend_elements(), title='Target')
ax.axhline(over_50['chol'].mean(), linestyle='--')
```

```python
plt.style.use('seaborn-whitegrid')
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'], cmap='winter');
ax.set(title='Heart disease and Max heart rate', xlabel='age', ylabel='thalach')
ax.legend(*scatter.legend_elements(), title='Target')
ax.axhline(over_50['thalach'].mean(), linestyle='--')
```

```python
plt.style.use('seaborn-whitegrid')
plt.style.use('ggplot')

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

scatter_age_chol = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['sex'], cmap='winter');
ax0.set(title='Heart disease and Cholestrol levels', xlabel='age', ylabel='chol')
ax0.set_xlim([50, 80])
ax0.legend(*scatter_age_chol.legend_elements(), title='Sex')
ax0.axhline(over_50['chol'].mean(), linestyle='dotted')


scatter_thalach_age = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['sex'], cmap='winter');
ax1.set(title='Heart disease and Max heart rate', xlabel='age', ylabel='max heart rate')
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
ax1.legend(*scatter_thalach_age.legend_elements(), title='Sex')
ax1.axhline(over_50['thalach'].mean(), linestyle='--');

fig.suptitle('Heart disease analysis', fontsize=24, fontweight='bold');
```
